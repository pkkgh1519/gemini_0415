#### 시스템 지시사항 및 파라미터 설정

```python
# system_instruction 으로 페르소나(Persona) 설정
from google.genai import types
from google import genai

client = genai.Client(vertexai=True, project="byounghwa-go", location="global")

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    config=types.GenerateContentConfig(
        system_instruction="당신은 초등학교 과학 선생님입니다",
    ),
    contents="하늘이 파란 이유를 간단히 설명해줘"
)
print(response.text)
```

```python
### 파라미터 설정
response = client.models.generate_content(
    model="gemini-3-flash-preview",
    config=types.GenerateContentConfig(
        system_instruction="당신은 기발하고 재미있는 아이디어를 내는 코미디 작가입니다",
        temperature=1.2,
        top_p=0.9,
        top_k=10,
        max_output_tokens=4096
    ),
    contents="관객이 재미있어할 만한 짧은 이야기를 2개만 만들어주세요"    
)
print(response.text)
```

#### 컨텍스트 캐싱(Context Caching)

```python
import datetime
from google import genai
from google.genai import types

# 1. 클라이언트 설정 (사용자님의 정보로 수정하세요)
PROJECT_ID = "byounghwa-go"
LOCATION = "global"

client = genai.Client(
    vertexai=True, 
    project=PROJECT_ID, 
    location=LOCATION
)

# 2. 1024 토큰 이상이 되도록 충분히 긴 가상의 텍스트 생성
# (한글 약 3,000자 이상으로 구성하여 1024 토큰 기준을 넉넉히 넘깁니다)
long_text_content = """
이 문서는 컨텍스트 캐싱 테스트를 위한 대규모 데이터 예시입니다. 
인공지능의 역사와 발전 과정에 대한 긴 내용을 담고 있습니다.
""" + ("인공지능(AI)은 인간의 지능을 컴퓨터로 구현하는 기술로, 1950년대 앨런 튜링의 논의부터 시작되었습니다. " * 50) + """
최근의 거대 언어 모델(LLM)은 트랜스포머 아키텍처를 기반으로 하며, 수십억 개의 파라미터를 통해 문맥을 이해합니다.
컨텍스트 캐싱은 이러한 긴 문맥을 매번 다시 계산하지 않고 저장해두는 혁신적인 기술입니다.
""" + ("데이터의 양이 충분해야 캐싱 효율이 극대화되며, 구글 Gemini는 1024 토큰 이상의 데이터에 대해 캐싱을 지원합니다. " * 50)

# 3. 토큰 수 미리 확인 (선택 사항이지만 안전함)
token_count_response = client.models.count_tokens(
    model="gemini-3-flash-preview",
    contents=long_text_content
)
print(f"입력된 데이터의 총 토큰 수: {token_count_response.total_tokens}")

# 4. 컨텍스트 캐시 생성 (1024 토큰 미만이면 여기서 에러가 나지만, 위에서 늘렸기에 안전합니다)
try:
    print("캐시 생성 중...")
    cache = client.caches.create(
        model="gemini-3-flash-preview",
        config=types.CreateCachedContentConfig(
            display_name="ai-history-cache",
            contents=[types.Content(
                role="user", 
                parts=[types.Part.from_text(text=long_text_content)]
            )],
            ttl="300s", # 5분 동안 유지
        )
    )
    print(f"캐시 생성 완료! 이름: {cache.name}")

    # 5. 생성된 캐시를 사용하여 질문하기
    response = client.models.generate_content(
        model="gemini-3-flash-preview",
        contents="위의 긴 내용 중에서 '인공지능의 시작'과 '컨텍스트 캐싱'의 이점이 무엇인지 요약해줘.",
        config=types.GenerateContentConfig(
            cached_content=cache.name
        )
    )

    print("\n--- 모델 응답 ---")
    print(response.text)

except Exception as e:
    print(f"오류 발생: {e}")
```

```python
# 응답 메타데이터 출력
usage = response.usage_metadata
print(f"전체 입력 토큰: {usage.prompt_token_count}")
print(f"캐시 적중 토큰: {usage.cached_content_token_count}") # 이 수치를 확인
print(f"생성된 토큰: {usage.candidates_token_count}")

if usage.cached_content_token_count > 0:
    hit_rate = (usage.cached_content_token_count / usage.prompt_token_count) * 100
    print(f"캐시 적중률: {hit_rate:.2f}% - 실제 캐싱이 적용되었습니다.")
else:
    print("캐시가 적용되지 않았습니다. 입력 토큰 수나 모델 설정을 확인하세요.")
```

```python
# 캐시 목록 확인 (메타데이터만 가져옴)
for c in client.caches.list():
    print(c)
```

```python
# 캐시 삭제 
client.caches.delete(name=cache.name)
print("캐시가 삭제되었습니다.")
```

#### 로깅(Logging)

```python
import vertexai
from vertexai.preview.generative_models import GenerativeModel

PROJECT_ID = "byounghwa-go"
LOCATION = "us-central1"   # "global"로 하면 오류남

vertexai.init(project=PROJECT_ID, location=LOCATION)

model = GenerativeModel("gemini-2.5-flash")  # 3.0 버전은 아직 정식으로 로깅 안됨

logging_config_result = model.set_request_response_logging_config(
    enabled=True,
    sampling_rate=1.0,                         # 100% 모든 요청 로깅
    bigquery_destination=f"bq://{PROJECT_ID}", # 자동 생성 경로 지정
    enable_otel_logging=True
)

print("Gemini 요청/응답 로깅이 성공적으로 활성화되었습니다.")   
```

```python
# 설정된 출력 URI 확인
output_uri = logging_config_result.logging_config.bigquery_destination.output_uri
print(f"로그 저장소 URI: {output_uri}")

response = model.generate_content("GCP의 로깅 서비스는 무엇인가요?")
print(f"모델 응답: {response.text}")

response = model.generate_content("GCP의 BigQuery 서비스는 무엇인가요?")
print(f"모델 응답: {response.text}")

# GCP Console 의 BigQuery 에서 테이블 확인해본다 : 요청/응답 후 일정 시간이 지나서야 테이블에 보여짐!!
```

```python
# 로깅 설정 삭제
logging_config_result = model.set_request_response_logging_config(
        enabled=False,
        sampling_rate=1.0,                        
        bigquery_destination=f"bq://{PROJECT_ID}",
    )
print("Gemini 요청/응답 로깅이 성공적으로 비활성화(삭제)되었습니다.")
```
